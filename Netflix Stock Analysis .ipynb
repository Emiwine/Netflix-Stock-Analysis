```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

from datetime import datetime  #we will show the data in the form of date and time 
```

```python
df = pd.read_csv('NFLX.csv')
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
df.head()
```

```python
sns.lineplot(x = df.index , y = df['Volume'] , label = 'Volume')
plt.title('volume of Stock vs Time')
```

### 2 - Netflix stock price - high , open , close

```python
df.plot(y = ['High' , 'Open','Close'] , title = 'Netflix Stock Price')
```

### 3 - Netflix stock price - Day , month and yearwise

```python
fig, (ax1,ax2,ax3) = plt.subplots(3 , figsize = (15 , 15)) #here we take the 3 subgraph and also give the fig size of the graph 
df.groupby(df.index.day).mean().plot(y = 'Volume', ax = ax1 , xlabel = 'Day')
df.groupby(df.index.month).mean().plot(y = 'Volume', ax = ax2 , xlabel = 'Month')
df.groupby(df.index.year).mean().plot(y = 'Volume', ax = ax3 , xlabel = 'Year')
```
